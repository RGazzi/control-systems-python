```python
!pip install control
!pip install matplotlib
!pip install numpy
!pip install spicy

import control as ctl
import matplotlib.pyplot as plt
import numpy as np
from control import TransferFunction

delta  = 5 #polo em 5
G1 = TransferFunction(1, [1, delta]) # denominador é 1 elevado a s, então denom= s + 5 (delta)

print('FT=/n', G1)
print('Polos =/n/n', ctl.poles(G1), '/n')
print('Zeros =/n/n', ctl.zeros(G1), '/n')

t = np.linspace(0, 5, 101) #vetor tempo para resposta ao degrau

#resposta ao degrau
plt.figure(1)
tout, yout = ctl.step_response(G1, t, X0=0) #função transferencia, tempo e condições iniciais
plt.plot(tout, yout, 'k', linewidth=1.5, label='resposta')
plt.legend(loc='best', shadow=True, framealpha=1)
plt.grid(alpha=0.3)
plt.xlabel('t')
plt.show()

#mapa de polos e zeros
plt.figure(2)
ctl.pzmap(G1, plot=True, grid=True) #pzmap = polos zero mapa
plt.title('Pole Zero Map')
plt.show()
```
