```python
#simbolic package
import sympy
sympy.init_printing()

#geração de simbolos
s, t = sympy.symbols('s, t') #s=var de laplace, t=tempo
```

```python
#função transformada de Laplace
def L(f):
  return sympy.laplace_transform(f, t, s, noconds = True)

#função transformada de Laplace inversa
def invL(F):
  sympy.inverse_laplace_transform(F, s, t)
```

```python
#transformada de Laplace da entrada
F = L(sympy.exp(-2*t)*2) #função exponencial -> e elevado a -2t ao quadrado
F
```

```python
#Função Transferência do sistema #relação entrada e saída
G = 1/(s**2 + 5*s + 4) #s ao quadrado + 5s + 4
G
```

```python
#resposta descrita no domínio de Laplace
X = G*F
X
```

```python
#Expansão em frações Parciais da resposta
X.apart(s)
```

```python
#resposta temporal via transformada inversa de Laplace
invL(X)
```

```python
#plotar a resposta encontrada
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 20, 1001) #vetor tempo
u = 2/3*(np.exp(-1*t))-1*(np.exp(-2*t))+1/3*(np.exp(-4*t)) #vetor resposta de acordo com o simbolico

#figura
plt.plot(t, u, 'k', linewidth=1.5, label = 'resposta')
plt.legend(loc='best', shadow=True, framealpha=1)
plt.grid(alpha=0.1)
plt.ylabel('posição (m)')
plt.xlabel('tempo (s)')
plt.show()
```
